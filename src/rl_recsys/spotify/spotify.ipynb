```python
import pandas as pd

file_path = '/home/aayush/git/rl_spotify/data/training_set/log_mini.csv'
file_path1  = '/home/aayush/git/rl_spotify/data/track_features/tf_mini.csv'
training_data = pd.read_csv(file_path)
features_data= pd.read_csv(file_path1)
```

```python


# merge the dataframes based on the track_id and track_id_clean columns
merged_df = pd.merge(training_data, features_data, left_on='track_id_clean', right_on='track_id')

```

```python


# sort the dataframe by the session_id column
sorted_df = merged_df.sort_values(by=['session_id','session_position'])
sorted_df = sorted_df.reset_index(drop=True)
```
